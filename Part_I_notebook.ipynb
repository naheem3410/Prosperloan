## Univariate Exploration

I will start by looking at the distribution of the variable of interest: BorrowerRate


Distribution of BorrowerRate

```python
#display basic statistics of the variable
df.BorrowerRate.describe()
```

```python
#generate tick numbers
def gen_tick(step,stop):
    ticks = []
    start = 0
    while start < stop:
        start += step
        ticks.append(start)
    return ticks
```

```python
#plot the histogram distribution of BorrowerRate
bins = np.arange(0.04, df.BorrowerRate.max()+0.01, 0.01)
#generate tick labels
ticks = gen_tick(0.02,0.5)
tick_labels = ["%.2f" % tick for tick in ticks]
plt.hist(data = df, x = 'BorrowerRate',bins=bins);
plt.xticks(ticks, tick_labels);
plt.xticks(rotation = 90);
plt.xlabel('Distribution of BorrowerRate');
```

The distribution is bimodal and slightly skewed to the right, with one large peak between 0.14 and 0.16, and a second large peak between 0.30 - 0.32

Distribution of EstimatedLoss

```python
#display basic statistics of the variable
df.EstimatedLoss.describe()
```

```python
#plot the distribution of the estimated loss
plt.hist(data = df, x = 'EstimatedLoss', bins = 100);
plt.xlabel('Distribution of EstimatedLoss');
```

The figure above shows that the distribution is highly skewed to the right. This will be corrected using a log transform

```python
#show basic descriptive statistics
df.EstimatedLoss.describe()
```

Make a log transformation of EstimatedLoss to improve interpretability

The distribution is unimodal and has its peak between 0.05 and 0.06

Distribution of IncomeRange

```python
#get the basic statistics of the variable
df.IncomeRange.describe()
```

```python
df.IncomeRange.value_counts()
```

```python
#convert income range to a ordered category data type
income_classes = ['Not employed','$0', '$1-24,999', '$25,000-49,999', '$50,000-74,999', '$75,000-99,999', '$100,000+']
#get current Pandas version
pd_ver = pd.__version__.split(".")
if (int(pd_ver[0]) > 0) or (int(pd_ver[1]) >= 21):#for version 0.21 or later
    incomeclasses = pd.api.types.CategoricalDtype(ordered = True, categories = income_classes)
    df['IncomeRange'] = df['IncomeRange'].astype(incomeclasses)
else:
    df['IncomeRange'] = df['IncomeRange'].astype('category', ordered = True, categories = income_classes)
```

```python
#plot the distribution of IncomeRange
base_color = sb.color_palette()[3]
sb.countplot(data=df, y='IncomeRange', color=base_color);
plt.xlabel('Distribution of borrowers income range');
```

The plot shows that borrowers within the income range of 50,000 - 74,999 has the highest frequency

## Bivariate Exploration

I would like to have an overview of the overall relationship between the variables I selected, before digging deeper

```python
#store the quantitative and qualitative variables for reference
numeric_vars = ['BorrowerRate', 'EstimatedLoss']
categoric_vars = ['EmploymentStatus', 'IsBorrowerHomeowner','IncomeRange']
```

```python
#make a correlation between the numerical variables on a heatmap
plt.figure(figsize = [8, 5])
sb.heatmap(df[numeric_vars].corr(), annot = True, fmt = '.3f',
           cmap = 'viridis_r', center = 0);
plt.show();
```

```python
#make a plot matrix - sample 500 data so that plots are clearer and execute faster
df_samp = df.head(500)
g = sb.PairGrid(data = df_samp, vars = numeric_vars);
g = g.map_diag(plt.hist, bins = 20);
g.map_offdiag(plt.scatter);
```

In the plots above, we can see some interesting trends in the variables. BorrowerRate and EstimatedLoss are very highly correlated positively. I will dig deep into this relationship later in the analysis

Relationship between BorrowerRate, Estimated Loss, and other qualitative variables

```python
#plot a boxplot
def boxgrid(x, y, **kwargs):
    base_color = sb.color_palette()[2]
    sb.boxplot(x, y, color = base_color)
    plt.xticks(rotation = 90);
```

```python
#plot the selected numerical variables against the selected categorical variables
plt.figure(figsize = [30, 30]);
g = sb.PairGrid(data = df, y_vars = ['BorrowerRate', 'EstimatedLoss'], x_vars = categoric_vars,size = 2, aspect = 1.5);
g.map(boxgrid);
plt.show();
```

The plots above reveal that a relationship exists between the borrowers rate and borrowers employment status as the borrowers employed received less median interest rate, while the ones that are not employed received high median interest rate. In that regard, borrowers that are home owners also received lower median interest rates compared to those that are not home owners. In the same way, borrowers that have high income received lesser median interest rate compared to those that received lower income. The same relationship also exists between estimated loss and other qualitative variables.

## Multivariate Exploration

To proceed with my analysis, I would like to know the interaction between BorrowerRate, EstimatedLoss and other categorical features

I will first plot the interaction between BorrowerRate, EstimatedLoss and EmploymentStatus

```python
#plot the interaction between BorrowerRate, EstimatedLoss and EmploymentStatus
g = sb.FacetGrid(data = df, hue = 'EmploymentStatus', size = 10);
g.map(plt.scatter, 'EstimatedLoss', 'BorrowerRate');
g.add_legend();
```

Looking at the plot above, one will likely think that there seems to be no relationship there, but a more closer look reveals that the borrowers that are employed still received a lower interest rates compared to those that are not employed.

```python
#plot the interaction between BorrowerRate, EstimatedLoss and IsBorrowerHomeowner
g = sb.FacetGrid(data = df, hue = 'IsBorrowerHomeowner', size = 10);
g.map(plt.scatter, 'EstimatedLoss', 'BorrowerRate');
g.add_legend();
```

Though, in the plot, borrowers that are home owners and those that do not appear to be distributed throughout the plot. But, looking closely will reveal that in some degree, borrowers that are homeowners still enjoy the benefit of receiving lower interest rates copared to those that do not own a home.

```python
#plot the interaction between BorrowerRate, EstimatedLoss and IncomeRange
g = sb.FacetGrid(data = df, hue = 'IncomeRange', size = 10);
g.map(plt.scatter, 'EstimatedLoss', 'BorrowerRate');
g.add_legend();
```

The plot shows that the population of borrowers that are unemployed and those that are earning between $1-24,999 are increasing as the BorrowerRate and EstimatedLoss are increasing. This further augment the fact that borrowers with high income range are likely to have lower interest rates assigned to them.
